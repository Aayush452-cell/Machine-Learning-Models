```python
from sklearn.covariance import EllipticEnvelope
```

```python
Outlier=EllipticEnvelope(contamination=0.1)
```

```python
features=df_x.loc[:,['MonthlyIncome']]
```

```python
features=df_x.loc[:,['MonthlyIncome']].values
outliers=[]
def detect_otliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=((i-mean)/std)
        if(np.abs(z_score)>threshold):
            outliers.append(i)
    return outliers
```

```python
print(detect_otliers(features))
```

```python
features=df_x.loc[:,['MonthlyRate']].values
```

```python
outliers=[]
def detect_outliers(data):
    mean=np.mean(data)
    std=np.std(data)
    threshold=3
    for i in data:
        z_score=(i-mean)/std
        if(np.abs(z_score)>threshold):
            outliers.append(i)
    return outliers
```

```python
print(detect_otliers(features))#no outliers are found.
```
